```python
import pandas as pd
import numpy as np
print("caca")
```

```python
excel_pf = pd.read_excel('stock_prices.xlsx')
print(excel_pf.head())
```

```python
excel_pf.rename(columns = {"symbol" :  "ticker"}, inplace=True)
print(excel_pf.columns)
```

```python
excel_pf = excel_pf.sort_values(by=["ticker", "date"])
print(excel_pf.head(20))
```

```python
print(excel_pf["ticker"].unique())
```

```python
symbols = excel_pf["ticker"].unique()
sample_20 = np.random.choice(symbols, size=20, replace=False)
excel_final = excel_pf[excel_pf["ticker"].isin(sample_20)]
print(excel_final["ticker"].unique())
```

```python
Liste_return = []
for i in excel_pf["ticker"].unique():
    ticker_df = excel_pf[excel_pf["ticker"] == i ]
    Liste_return.append((ticker_df["close"].iloc[-1] - ticker_df["close"].iloc[0])/ticker_df["close"].iloc[0])
print(Liste_return)
```

```python
arr = np.array(Liste_return)
top10_idx = np.argsort(arr)[-10:][::-1]   
print(top10_idx)
```

```python
Liste_ticker_max = [excel_pf["ticker"].unique()[i] for i in top10_idx]
print(Liste_ticker_max)
print(type(Liste_ticker_max))
```

```python
sample_10 = np.random.choice(symbols, size=10, replace=False)
print(type(sample_10))
```

```python
Liste_ticker_excel = list(sample_10) + Liste_ticker_max
print(Liste_ticker_excel)
```

```python
excel_ticker = excel_pf[excel_pf["ticker"].isin(Liste_ticker_excel)]
excel_ticker.to_excel("stock_prices_clean1.xlsx", index=False)
```

```python
excel_pf["return"] = excel_pf.groupby("ticker")["close"].pct_change()
df_20 = excel_pf[excel_pf["ticker"].isin(Liste_ticker_excel)]
df_20.dropna(inplace=True)

returns_wide = df_20.pivot(index="date", columns="ticker", values="return")
print(returns_wide.head())
corr_matrix = returns_wide.corr()
print(corr_matrix)
```

```python
corr_matrix.to_excel("covariance_matrix.xlsx", sheet_name="Covariances")
```
